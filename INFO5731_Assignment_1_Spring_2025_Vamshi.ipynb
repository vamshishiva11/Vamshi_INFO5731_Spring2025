```python
import math

def evaluate_expression(expression):
    try:
        # Evaluating the mathematical expression
        return eval(expression)
    except ZeroDivisionError:
        print("Error: Division by zero is not allowed.")
        return None
    except Exception as e:
        print(f"Error: {e}")
        return None

def calculator():
    print("Simple Command-Line Calculator")
    print("Supported operations: +, -, *, /, ** (Exponentiation), sqrt (Square root), % (Modulus)")
    print("Type 'exit' to quit.")

    while True:
        # Getting user input
        expression = input("\nEnter an expression: ")

        if expression.lower() == 'exit':
            print("Exiting the calculator.")
            break

        # Handling square root operation
        expression = expression.replace("sqrt", "math.sqrt")

        # Handle the evaluation
        result = evaluate_expression(expression)

        if result is not None:
            print(f"Result: {result}")

if __name__ == "__main__":
    calculator()
```

The file contains statistics about housing in the city of Denton from 2008 to 2014. Write a program to implement the following questions.

                              Year	Value
	Vacant Housing Units	  -----   -----

(d) (5 pts) Calculate the harmonic mean of the total housing units and print it out. You can find the information about harmonic mean here: https://ncalculators.com/statistics/harmonic-mean-calculator.htm

```python
import pandas as pd
import numpy as np

# Read the CSV file
df = pd.read_csv('/content/Assignment1_denton_housing.csv')

# Create a pivot table for easier analysis
df_pivot = df.pivot(index='year', columns='title_field', values='value')

# (a) Calculate year-over-year differences in Occupied Housing Units
print("(a) Year-over-year differences in Occupied Housing Units:")
differences = []
years = sorted(df_pivot.index.unique())

for i in range(len(years)-1):
    curr_year = df_pivot.loc[years[i], 'Occupied Housing Units']
    next_year = df_pivot.loc[years[i+1], 'Occupied Housing Units']
    diff = next_year - curr_year
    differences.append(diff)
    print(f"{years[i]}-{years[i+1]}: {diff}")

print("\nDifferences in ascending order:", sorted(differences))

# (b) Calculate percentages of vacant and occupied units
print("\n(b) Housing unit percentages:")
print("Year\tVacant Housing Units\tOccupied Housing Units")
for year in years:
    total = df_pivot.loc[year, 'Total Housing Units']
    vacant_pct = (df_pivot.loc[year, 'Vacant Housing Units'] / total) * 100
    occupied_pct = (df_pivot.loc[year, 'Occupied Housing Units'] / total) * 100
    print(f"{year}\t{vacant_pct:.1f}%\t\t\t{occupied_pct:.1f}%")

# (c) Find years with highest vacant and occupied units
print("\n(c) Peak years:")
max_vacant_year = df_pivot['Vacant Housing Units'].idxmax()
max_occupied_year = df_pivot['Occupied Housing Units'].idxmax()
print("\t\t\tYear\tValue")
print(f"Vacant Housing Units\t{max_vacant_year}\t{df_pivot.loc[max_vacant_year, 'Vacant Housing Units']:.0f}")
print(f"Occupied Housing Units\t{max_occupied_year}\t{df_pivot.loc[max_occupied_year, 'Occupied Housing Units']:.0f}")

# (d) Calculate harmonic mean of total housing units
total_units = df_pivot['Total Housing Units']
harmonic_mean = len(total_units) / sum(1/x for x in total_units)
print(f"\n(d) Harmonic mean of total housing units: {harmonic_mean:.2f}")
```

# Question 6 (20 points).
Develop a advance Python program to manage inventory records for a prodcut managing company. The system needs functions for adding, deleting and changing products and basic search options. Your system should use CSV files to automate restocking and produce reports that show item values alongside category breakdowns. Manage edge cases and corrupted CSV files. Finally maintaine 50 products of different categories in inventory.

```python
import csv
import os
from datetime import datetime
import pandas as pd

class InventorySystem:
    def __init__(self, filename="inventory.csv", backup_dir="backups/"):
        """Initialize inventory system with file handling and backup support"""
        self.filename = filename
        self.backup_dir = backup_dir
        self.headers = ["ID", "Name", "Category", "Quantity", "Price", "Last Updated"]
        self.valid_categories = {"Electronics", "Clothing", "Food", "Books", "Tools"}

        # Create backup directory if it doesn't exist
        os.makedirs(backup_dir, exist_ok=True)

        # Create or load inventory file
        if not os.path.exists(filename):
            self.create_initial_inventory()
        else:
            self.validate_csv_structure()

    def create_initial_inventory(self):
        """Create initial inventory with 50 sample products"""
        sample_products = []
        categories = ["Electronics", "Clothing", "Food", "Books", "Tools"]
        products = [
            ("Laptop", 1299.99), ("Smartphone", 899.99), ("Headphones", 199.99),
            ("T-Shirt", 29.99), ("Jeans", 79.99), ("Jacket", 129.99),
            ("Rice", 12.99), ("Coffee", 15.99), ("Pasta", 5.99),
            ("Python Book", 49.99), ("SQL Guide", 39.99), ("Java Manual", 44.99),
            ("Hammer", 19.99), ("Screwdriver", 9.99), ("Drill", 149.99)
        ]

        for i in range(1, 51):
            prod_idx = (i - 1) % len(products)
            cat_idx = (i - 1) % len(categories)

            product = {
                "ID": f"P{i:03d}",
                "Name": f"{products[prod_idx][0]} {i}",
                "Category": categories[cat_idx],
                "Quantity": 100 + (i * 2),
                "Price": products[prod_idx][1],
                "Last Updated": datetime.now().strftime("%Y-%m-%d")
            }
            sample_products.append(product)

        self.write_inventory(sample_products)

    def validate_csv_structure(self):
        """Validate CSV file structure and create backup if corrupted"""
        try:
            df = pd.read_csv(self.filename)
            if not all(header in df.columns for header in self.headers):
                raise ValueError("Invalid CSV structure")
        except Exception as e:
            print(f"Error in CSV file: {e}")
            self.restore_from_backup()

    def read_inventory(self):
        """Read inventory from CSV file"""
        try:
            with open(self.filename, 'r', newline='') as file:
                reader = csv.DictReader(file)
                return list(reader)
        except Exception as e:
            print(f"Error reading inventory: {e}")
            return []

    def write_inventory(self, products):
        """Write inventory to CSV file"""
        try:
            with open(self.filename, 'w', newline='') as file:
                writer = csv.DictWriter(file, fieldnames=self.headers)
                writer.writeheader()
                writer.writerows(products)
            return True
        except Exception as e:
            print(f"Error writing inventory: {e}")
            return False

    def backup_inventory(self):
        """Create timestamped backup of inventory"""
        timestamp = datetime.now().strftime("%Y%m%d_%H%M%S")
        backup_file = os.path.join(self.backup_dir, f"inventory_backup_{timestamp}.csv")
        try:
            with open(self.filename, 'r') as source, open(backup_file, 'w', newline='') as target:
                target.write(source.read())
            return True
        except Exception as e:
            print(f"Backup failed: {e}")
            return False

    def restore_from_backup(self):
        """Restore inventory from most recent backup"""
        try:
            backup_files = sorted([f for f in os.listdir(self.backup_dir) if f.startswith("inventory_backup_")])
            if not backup_files:
                print("No backup files found")
                return False

            latest_backup = os.path.join(self.backup_dir, backup_files[-1])
            with open(latest_backup, 'r') as source, open(self.filename, 'w') as target:
                target.write(source.read())
            return True
        except Exception as e:
            print(f"Error restoring from backup: {e}")
            return False

    def add_product(self, name, category, quantity, price):
        """Add new product with validation"""
        try:
            if not name or len(name) < 3:
                raise ValueError("Invalid product name")
            if category not in self.valid_categories:
                raise ValueError("Invalid category")
            if not isinstance(quantity, int) or quantity < 0:
                raise ValueError("Invalid quantity")
            if not isinstance(price, (int, float)) or price <= 0:
                raise ValueError("Invalid price")
            products = self.read_inventory()
            new_id = f"P{len(products) + 1:03d}"

            new_product = {
                "ID": new_id,
                "Name": name,
                "Category": category,
                "Quantity": quantity,
                "Price": price,
                "Last Updated": datetime.now().strftime("%Y-%m-%d")
            }

            self.backup_inventory()
            products.append(new_product)
            self.write_inventory(products)
            return True
        except Exception as e:
            print(f"Error adding product: {e}")
            return False

    def delete_product(self, product_id):
        """Delete product with validation and backup"""
        try:
            self.backup_inventory()
            products = self.read_inventory()
            original_len = len(products)
            products = [p for p in products if p["ID"] != product_id]

            if len(products) == original_len:
                raise ValueError("Product not found")

            self.write_inventory(products)
            return True
        except Exception as e:
            print(f"Error deleting product: {e}")
            return False

    def update_product(self, product_id, **kwargs):
        """Update product with validation"""
        try:
            self.backup_inventory()
            products = self.read_inventory()
            product_found = False

            for product in products:
                if product["ID"] == product_id:
                    product_found = True
                    for key, value in kwargs.items():
                        if key == "Category" and value not in self.valid_categories:
                            raise ValueError("Invalid category")
                        if key == "Quantity" and (not isinstance(value, int) or value < 0):
                            raise ValueError("Invalid quantity")
                        if key == "Price" and (not isinstance(value, (int, float)) or value <= 0):
                            raise ValueError("Invalid price")
                        product[key] = value
                    product["Last Updated"] = datetime.now().strftime("%Y-%m-%d")

            if not product_found:
                raise ValueError("Product not found")

            self.write_inventory(products)
            return True
        except Exception as e:
            print(f"Error updating product: {e}")
            return False

    def search_products(self, **kwargs):
        """Search products with flexible criteria"""
        try:
            products = self.read_inventory()
            results = products

            for key, value in kwargs.items():
                if key == "price_range":
                    min_price, max_price = value
                    if min_price > max_price:
                        raise ValueError("Invalid price range")
                    results = [p for p in results if float(p["Price"]) >= min_price and float(p["Price"]) <= max_price]
                elif key == "quantity_below":
                    results = [p for p in results if int(p["Quantity"]) < value]
                else:
                    results = [p for p in results if str(p.get(key, "")).lower() == str(value).lower()]

            return results
        except Exception as e:
            print(f"Error searching products: {e}")
            return []

    def generate_report(self, report_type="summary"):
        """Generate various types of inventory reports"""
        try:
            products = self.read_inventory()

            if report_type == "summary":
                total_value = sum(float(product["Price"]) * int(product["Quantity"]) for product in products)
                category_breakdown = {}
                for product in products:
                    cat = product["Category"]
                    if cat not in category_breakdown:
                        category_breakdown[cat] = 0
                    category_breakdown[cat] += float(product["Price"]) * int(product["Quantity"])

                return {
                    "total_value": total_value,
                    "category_breakdown": category_breakdown,
                    "total_products": len(products)
                }

            elif report_type == "low_stock":
                return [product for product in products if int(product["Quantity"]) < 50]

            elif report_type == "high_value":
                return sorted(products, key=lambda x: float(x["Price"]) * int(x["Quantity"]), reverse=True)[:10]

        except Exception as e:
            print(f"Error generating report: {e}")
            return None

# Test the system with edge cases
if __name__ == "__main__":
    inventory = InventorySystem()

    print("\nTesting edge cases:")

    # Edge case 1: Add product with negative quantity
    print("\n1. Adding product with negative quantity:")
    inventory.add_product("Test Product", "Electronics", -5, 99.99)

    # Edge case 2: Add product with invalid category
    print("\n2. Adding product with invalid category:")
    inventory.add_product("Test Product", "Invalid", 10, 99.99)

    # Edge case 3: Update non-existent product
    print("\n3. Updating non-existent product:")
    inventory.update_product("P999", quantity=50)

    # Edge case 4: Search with price range
    print("\n4. Searching products in price range:")
    results = inventory.search_products(price_range=(50, 100))
    print(f"Found {len(results)} products in price range")

    # Edge case 5: Add product with zero price
    print("\n5. Adding product with zero price:")
    inventory.add_product("Free Product", "Electronics", 10, 0)

    # Edge case 6: Add product with empty name
    print("\n6. Adding product with empty name:")
    inventory.add_product("", "Electronics", 10, 99.99)

    # Edge case 7: Add product with extremely large quantity
    print("\n7. Adding product with large quantity:")
    inventory.add_product("Bulk Item", "Electronics", 1000000, 99.99)

    # Edge case 8: Update product with invalid price
    print("\n8. Updating product with invalid price:")
    inventory.update_product("P001", Price=-10)

    # Edge case 9: Delete non-existent product
    print("\n9. Deleting non-existent product:")
    inventory.delete_product("P999")

    # Edge case 10: Search with invalid criteria
    print("\n10. Searching with invalid criteria:")
    results = inventory.search_products(invalid_field="test")

    # Generate and display report
    print("\nGenerating inventory report:")
    report = inventory.generate_report()
    if report:
        print(f"\nTotal Products: {report['total_products']}")
        print(f"Total Inventory Value: ${report['total_value']:,.2f}")
        print("\nCategory Breakdown:")
        for category, value in report['category_breakdown'].items():
            print(f"{category}: ${value:,.2f}")
    else:
        print("Error generating report")

```

```python
from datetime import datetime, timedelta
import csv
import os
import random
import time

class FlightBookingSystem:
    def __init__(self, flights_file="flights.csv", reservations_file="reservations.csv"):
        """Initialize the flight booking system with file handling"""
        self.flights_file = flights_file
        self.reservations_file = reservations_file
        self.valid_cities = {"New York", "London", "Tokyo", "Paris",
                             "Dubai", "Singapore", "Sydney", "Mumbai",
                             "Beijing", "Toronto"}

        # Create files if they don't exist
        if not os.path.exists(flights_file):
            self.initialize_flights()
        if not os.path.exists(reservations_file):
            self.create_reservations_file()

    def initialize_flights(self):
        """Create initial set of 50 flights"""
        headers = ["FlightID", "DepartureCity", "ArrivalCity", "DepartureTime",
                   "TotalSeats", "AvailableSeats", "Price", "Status"]

        flights = []
        cities = list(self.valid_cities)

        for i in range(1, 51):
            departure = random.choice(cities)
            arrival = random.choice([c for c in cities if c != departure])

            departure_time = (datetime.now() + timedelta(
                days=random.randint(1, 30),
                hours=random.randint(0, 23),
                minutes=random.randint(0, 59)
            )).strftime("%Y-%m-%d %H:%M")

            flight = {
                "FlightID": f"FL{i:03d}",
                "DepartureCity": departure,
                "ArrivalCity": arrival,
                "DepartureTime": departure_time,
                "TotalSeats": 200,
                "AvailableSeats": 200,
                "Price": round(random.uniform(200, 1000), 2),
                "Status": "Scheduled"
            }
            flights.append(flight)

        with open(self.flights_file, 'w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=headers)
            writer.writeheader()
            writer.writerows(flights)

    def create_reservations_file(self):
        """Create reservations file with headers"""
        headers = ["ReservationID", "FlightID", "PassengerName", "Seats",
                   "BookingTime", "Status"]
        with open(self.reservations_file, 'w', newline='') as file:
            writer = csv.DictWriter(file, fieldnames=headers)
            writer.writeheader()

    def search_flights(self, departure, arrival, date=None, max_price=None):
        """Search flights with multiple criteria"""
        try:
            if departure not in self.valid_cities or arrival not in self.valid_cities:
                raise ValueError("Invalid city specified")
            if departure == arrival:
                raise ValueError("Departure and arrival cities cannot be the same")

            with open(self.flights_file, 'r', newline='') as file:
                flights = list(csv.DictReader(file))

            results = []
            for flight in flights:
                if (flight["DepartureCity"] == departure and flight["ArrivalCity"] == arrival
                    and flight["Status"] == "Scheduled"):

                    if date:
                        flight_date = datetime.strptime(flight["DepartureTime"], "%Y-%m-%d %H:%M").date()
                        if flight_date != date:
                            continue

                    if max_price and float(flight["Price"]) > max_price:
                        continue

                    results.append(flight)

            return results

        except Exception as e:
            print(f"Error searching flights: {e}")
            return []

    def make_reservation(self, flight_id, passenger_name, seats=1):
        """Make a new reservation with validation and unique ID generation"""
        try:
            if not passenger_name or len(passenger_name) < 2:
                raise ValueError("Invalid passenger name")
            if not isinstance(seats, int) or seats < 1:
                raise ValueError("Invalid number of seats")

            # Read current flights
            with open(self.flights_file, 'r', newline='') as file:
                flights = list(csv.DictReader(file))
            headers = flights[0].keys()

            # Find and update flight
            flight_found = False
            for flight in flights:
                if flight["FlightID"] == flight_id:
                    flight_found = True

                    departure_time = datetime.strptime(flight["DepartureTime"], "%Y-%m-%d %H:%M")
                    if departure_time < datetime.now():
                        raise ValueError("Flight has already departed")

                    available_seats = int(flight["AvailableSeats"])
                    if available_seats < seats:
                        raise ValueError("Not enough seats available")

                    flight["AvailableSeats"] = str(available_seats - seats)
                    break

            if not flight_found:
                raise ValueError("Flight not found")

            # Create unique reservation ID
            timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
            unique_id = random.randint(1000, 9999)
            time.sleep(0.001)  # Ensure unique timestamps
            reservation_id = f"RES{timestamp}_{unique_id}"

            # Verify reservation ID uniqueness
            existing_ids = set()
            if os.path.exists(self.reservations_file):
                with open(self.reservations_file, 'r', newline='') as file:
                    reader = csv.DictReader(file)
                    existing_ids = {row["ReservationID"] for row in reader}

            while reservation_id in existing_ids:
                unique_id = random.randint(1000, 9999)
                reservation_id = f"RES{timestamp}_{unique_id}"

            reservation = {
                "ReservationID": reservation_id,
                "FlightID": flight_id,
                "PassengerName": passenger_name,
                "Seats": seats,
                "BookingTime": datetime.now().strftime("%Y-%m-%d %H:%M"),
                "Status": "Confirmed"
            }

            # Save updated flights
            with open(self.flights_file, 'w', newline='') as file:
                writer = csv.DictWriter(file, fieldnames=headers)
                writer.writeheader()
                writer.writerows(flights)

            # Save reservation
            write_header = not os.path.exists(self.reservations_file) or os.path.getsize(self.reservations_file) == 0
            with open(self.reservations_file, 'a', newline='') as file:
                writer = csv.DictWriter(file, fieldnames=reservation.keys())
                if write_header:
                    writer.writeheader()
                writer.writerow(reservation)

            print(f"Reservation {reservation_id} created successfully")
            return reservation_id

        except Exception as e:
            print(f"Error making reservation: {e}")
            return None

    def cancel_reservation(self, reservation_id):
        """Cancel a reservation and update flight seats"""
        try:
            # Read reservations
            with open(self.reservations_file, 'r', newline='') as file:
                reservations = list(csv.DictReader(file))
            if not reservations:
                raise ValueError("No reservations found")
            res_headers = reservations[0].keys()

            # Find reservation
            reservation_found = False
            flight_id = None
            seats_to_return = 0

            for reservation in reservations:
                if reservation["ReservationID"] == reservation_id:
                    reservation_found = True
                    if reservation["Status"] == "Cancelled":
                        raise ValueError("Reservation is already cancelled")

                    flight_id = reservation["FlightID"]
                    seats_to_return = int(reservation["Seats"])
                    reservation["Status"] = "Cancelled"
                    break

            if not reservation_found:
                raise ValueError("Reservation not found")

            # Update flight seats
            with open(self.flights_file, 'r', newline='') as file:
                flights = list(csv.DictReader(file))
            flight_headers = flights[0].keys()

            for flight in flights:
                if flight["FlightID"] == flight_id:
                    departure_time = datetime.strptime(flight["DepartureTime"], "%Y-%m-%d %H:%M")
                    if departure_time < datetime.now():
                        raise ValueError("Cannot cancel - flight has already departed")

                    flight["AvailableSeats"] = str(int(flight["AvailableSeats"]) + seats_to_return)
                    break

            # Save updated data
            with open(self.flights_file, 'w', newline='') as file:
                writer = csv.DictWriter(file, fieldnames=flight_headers)
                writer.writeheader()
                writer.writerows(flights)

            with open(self.reservations_file, 'w', newline='') as file:
                writer = csv.DictWriter(file, fieldnames=res_headers)
                writer.writeheader()
                writer.writerows(reservations)

            print(f"Reservation {reservation_id} cancelled successfully")
            return True

        except Exception as e:
            print(f"Error cancelling reservation: {e}")
            return False

    def generate_report(self):
        """Generate booking report"""
        try:
            with open(self.flights_file, 'r', newline='') as file:
                flights = list(csv.DictReader(file))

            with open(self.reservations_file, 'r', newline='') as file:
                reservations = list(csv.DictReader(file))

            total_revenue = 0
            total_bookings = 0
            route_stats = {}

            for reservation in reservations:
                if reservation["Status"] == "Confirmed":
                    flight_id = reservation["FlightID"]
                    seats = int(reservation["Seats"])

                    for flight in flights:
                        if flight["FlightID"] == flight_id:
                            total_revenue += float(flight["Price"]) * seats
                            total_bookings += seats

                    route = f"{flight['DepartureCity']}-{flight['ArrivalCity']}"
                    route_stats[route] = route_stats.get(route, 0) + seats

            return {
                "total_revenue": total_revenue,
                "total_bookings": total_bookings,
                "route_statistics": route_stats
            }

        except Exception as e:
            print(f"Error generating report: {e}")
            return None

# Test the system
if __name__ == "__main__":
    booking_system = FlightBookingSystem()

    print("\nTesting edge cases:")

    # Edge case 1: Search invalid route
    print("\n1. Searching invalid route:")
    results = booking_system.search_flights("Mars", "Venus")
    print(f"Found {len(results)} flights")

    # Edge case 2: Search same departure and arrival
    print("\n2. Searching same departure and arrival:")
    results = booking_system.search_flights("London", "London")
    print(f"Found {len(results)} flights")

    # Edge case 3: Book with invalid passenger name
    print("\n3. Booking with invalid passenger name:")
    res_id = booking_system.make_reservation("FL001", "", 1)

    # Edge case 4: Book more seats than available
    print("\n4. Booking more seats than available:")
    res_id = booking_system.make_reservation("FL001", "John Doe", 201)

    # Edge case 5: Cancel non-existent reservation
    print("\n5. Cancelling non-existent reservation:")
    result = booking_system.cancel_reservation("RES99999999")

    # Edge case 6: Book with negative seats
    print("\n6. Booking with negative seats:")
    res_id = booking_system.make_reservation("FL001", "John Doe", -1)

    # Edge case 7: Search with future date
    print("\n7. Searching with future date:")
    future_date = datetime.now().date() + timedelta(days=60)
    results = booking_system.search_flights("London", "Paris", date=future_date)
    print(f"Found {len(results)} flights")

    # Edge case 8: Book with very long passenger name
    print("\n8. Booking with very long passenger name:")
    long_name = "A" * 100
    res_id = booking_system.make_reservation("FL001", long_name, 1)

    # Edge case 9: Search with very low max price
    print("\n9. Searching with very low max price:")
    results = booking_system.search_flights("London", "Paris", max_price=1)
    print(f"Found {len(results)} flights")

    # Edge case 10: Multiple bookings for same flight
    print("\n10. Multiple bookings for same flight:")
    for _ in range(3):  # Reduced to 3 bookings to stay within seat limit
        res_id = booking_system.make_reservation("FL001", "Test User", 50)

    # Generate and display report
    print("\nGenerating booking report:")
    report = booking_system.generate_report()
    if report:
        print(f"\nTotal Revenue: ${report['total_revenue']:,.2f}")
        print(f"Total Bookings: {report['total_bookings']}")
        print("\nPopular Routes:")
        for route, bookings in report['route_statistics'].items():
            print(f"{route}: {bookings} passengers")

```
